```python
import pandas as pd
```

```python
students = pd.read_excel("sdu_registration.xlsx",) #read excel file
students
```

```python
num_of_stud = students.groupby(["YEAR","EDU_LEVEL","CLASS"]).size() #number of students in each year by level of education
num_of_stud
```

```python
cour = students.loc[:,["EDU_LEVEL","CLASS","DERS_KOD","COURSE_TITLE"]]
AVE = cour.groupby("DERS_KOD").count().CLASS.mean()
AVE = int(AVE)
print(AVE) #courses average
```

```python
courses=cour.groupby("DERS_KOD").count()
courses.loc[courses['CLASS']>AVE] #courses more than average courses
```

```python
grades = students.loc[:,["SECTION","TEACHER","COURSE_TITLE"]]
teacher=grades.groupby("TEACHER").count()
teachers_list=teacher.loc[teacher['SECTION']>200] #teachers who more than 200 times fill sections for grades
teachers_list
```

```python
courses = students.loc[:,["COURSE_TITLE","SPECIALITY"]]
edu_prog=courses.groupby("SPECIALITY").count() #number of courses in each educational program 
edu_prog
```

```python
%matplotlib inline
edu_prog.plot.bar() #bar plot which shows which educational program has more courses
```
