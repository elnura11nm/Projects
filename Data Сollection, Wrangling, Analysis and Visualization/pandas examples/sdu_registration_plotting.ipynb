```python
import pandas as pd
import matplotlib.pyplot as plt
```

import libraries pandas and matplot for read the data, analyze them and draw plot

```python
students=pd.read_excel('sdu_registration.xlsx')
students
```

read data from *sdu_registration* excel file and create data frame *students*

```python
courses=students.loc[:,['STUD_ID','DERS_KOD']]
cours=courses.groupby('DERS_KOD').count()
plt.hist(cours,bins=5,histtype='stepfilled')
plt.show()
```

```python
cours_by_stud=pd.cut(cours['STUD_ID'], bins=5)
cours_by_stud
```

1-In plot presented dataframe *cours* which grouped by courses and divided into 5, in dataframe *cours_by_stud* divided by students into 5 categories 

```python
teacher_df = students[['TEACHER']].groupby('TEACHER').size().reset_index(name='counts')
teacher_df = teacher_df.sort_values(by=['counts'])
teacher_df.plot.bar(x='TEACHER', y='counts', rot=0)
```

2-Create *teacher_df* dataframe by using loc(when you can access specific coloumn by using index or coloumn label), group it and sort by counts. After that draw plot by using prepared dataframe

```python
TC_df = students.groupby(['TEACHER','CIPHER']).size().reset_index(name='counts')
plt.scatter(TC_df['CIPHER'], TC_df['TEACHER'], s=TC_df['counts'])
```

3-Plot shows teachers, cipher with size by students count 

```python
level_df =  students.groupby('EDU_LEVEL').TEACHER.nunique().reset_index(name='counts')
level_df.plot.barh(x='EDU_LEVEL', y='counts', rot=0)
```
